```python
df = pd.read_csv("datos.csv")

# Obtener porcentaje cobrado de Tax
df['Tasa_tax'] = df['Tax Booking']/(df['Precio x Noche']*df['N° Noches'])
df.fillna(0, inplace=True)
df['Tasa_tax'] = np.round(df['Tasa_tax'], 2)
df.drop(axis=1, columns='Tax Booking', inplace=True)

df
```

```python
df['Medio de Pago'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Inicializamos el OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)  # drop='first' elimina una columna dummy para evitar colinealidad

# Aplicamos el encoder a las 3 columnas categóricas
encoded_columns = encoder.fit_transform(df[['Canal de Venta', 'Medio de Pago']])

# Obtener los nombres de las nuevas columnas
encoded_column_names = encoder.get_feature_names_out(['Canal de Venta', 'Medio de Pago'])

# Crear un DataFrame con las columnas codificadas
encoded_df = pd.DataFrame(encoded_columns, columns=encoded_column_names)

# Concatenar el DataFrame codificado con el original (sin las columnas categóricas)
df_final = pd.concat([df.drop(['Canal de Venta', 'Medio de Pago'], axis=1), encoded_df], axis=1)

# Imprimir el resultado
df_final
```
